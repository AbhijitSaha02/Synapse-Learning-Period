```python
#import all the prerequisites like tensorflow and keras

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
#create a model using sequential
#add dense layers and activation function

model = Sequential([
    Dense(units = 32, input_dim = 16, activation = 'relu'),
    Dense(units = 16, activation = 'relu'),
    Dense(units = 1, activation = "sigmoid")
])
```

```python
#import your dataset

import pandas as pd

dataset = pd.read_csv('/content/gdrive/MyDrive/synapse_w2/titanic_train.csv')
```

```python
dataset.head()
```

```python
#create the features and labels of the dataset
#run the code

feature_name = ["Age","Fare", "Family_members", "Pclass_1", "Pclass_2", "Pclass_3", 
                "Sex_female", "Sex_male", "Title_Master", "Title_Miss", 
                "Title_Mr", "Title_Mrs", "Title_Ms", "Embarked_C", "Embarked_Q", 
                "Embarked_S"]
X = dataset[feature_name]
Y = dataset["Survived"]
```

```python
#Split the dataset into train,test dataset

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, 
                                                    random_state = 20201022)
```

```python
#compile your model using optimizers like "SGD","ADAM"
#use model.fit to train 

model.compile(optimizer = "SGD", loss = 'binary_crossentropy',
              metrics = ["binary_accuracy"])
model_result = model.fit(X, Y, batch_size = 50, epochs = 200, validation_split = 0.2, 
                         shuffle = True, verbose = 2)
```

```python
#use model.predict to predict the survival rate on the test dataset and print the predictions

predict = model.predict(X_test)
predict
```
